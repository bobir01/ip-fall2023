```python
#@title Task#1 {display-mode : "form"}

"""
1. Read an image, subtract 5 from all the intensities using for loop, write it with the name imageSub5.jpg.
You may use “imwrite” function of OpenCV for such purpose.
"""

def perform_task1(img: np.ndarray):
    org_img = img.copy()
    for i in range(len(img)):
        for j in range(len(img[i])):
            for k in range(len(img[i][j])):
                img[i, j, k] -= 5
    NUM_ROWS = 1
    IMGs_IN_ROW = 2
    f, ax = plt.subplots(NUM_ROWS, IMGs_IN_ROW, figsize=(16,6))

    ax[0].imshow(org_img[...,::-1]) # due to brg to rgb
    ax[1].imshow(img[...,::-1])

    ax[0].set_title('Before: ')
    ax[1].set_title('After subtracting 5:')

    title = 'Task1 results'
    f.suptitle(title, fontsize=16)
    plt.tight_layout()
    plt.show()

perform_task1(cv2.imread('cau.png'))

```
